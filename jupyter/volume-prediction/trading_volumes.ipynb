# Trading Volumes
How much does market interest in a security vary?

```python
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df_goog = pd.read_csv("HistoricalQuotes _flrs.csv", thousands=',')
```

```python
plt.hist(df_goog["volume"], normed=True, bins=30)
plt.xscale('log')
```

```python
quiet_days = df_goog[df_goog["volume"] < 100000]
quiet_days
```

```python
busy_days = df_goog[df_goog["volume"] > 2000000]
busy_days
```

```python
print("Highest volume was " + str(np.max(df_goog["volume"])) + ", minimum was " + str(np.min(df_goog["volume"])))
```
